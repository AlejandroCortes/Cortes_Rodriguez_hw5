##### FILTROS PASA ALTAS Y PASA BAJAS EN IMAGENES

A continuación se realiza el filtrado de 4 señales bidimencionales (Imágenes) por medio de un filtro pasa bajas y otro pasa altas (frecuencias), usando la transformada de Fourier bidimensional en Python.

##### El procedimiento realizado se describe acontinuación:

1) Se carga la imagen incial y se mira el numero de capas que tiene cada imagen y de que tamaño son.

2) Se realiza la Transformada de fourier de la imagen.

3) Luego se divide la o las matrices en cuatro partes iguales para posteriormente cambiar su orden, como se muestra en la Figura 1.

4) Para aplicar cada uno de los filtros seleccionamos un circulo con centro en la mitad de la matriz. Si el filtro es pasa bajas los los valores que se encuentran ubicados dentro del circulo se convierten en 0. Pero, si el circulo es pasa altas los valores que estan fuera del cirulo se convierten en 0.

5) Para regresar a la nueva imagen tenemos que volver a organizar la matriz en las posiciones iniciales, es decir regresar el punto 3.

6) Aplicamos Transformada Inversa de Fourier a la matriz del punto 5.

7) Graficamos la imagen final.

## FILTROS DE LA IMAGEN:  20_popc_cho009-1.tif

```python
#Punto 1
#Carga y muestra la imagen 20_popc_cho009-1.tif
imagen1 = pylab.imread("Imagenes/20_popc_cho009-1.tif")
fig = figure(1, figsize=(5,5))
imshow(imagen1[:,:], cmap = cm.Greys_r)
plt.title('$IMAGEN$'' ''$ORIGINAL$', fontsize=20)
```

```python
#Punto 2
#Usando la libreria de la Transformada de Fourier(TF) se halla la TF bidimensional 
#para la imagen (señal)
T1 = fft2(imagen1) 
#Imprime el tamaño de la matriz que es el resultado de la TF bidimensional
print np.shape(T1)
```

Esta imagen tiene 560 pixeles de alto, 512 de largo y consta de 1 capa

```python
#Punto 3
#Dividimos la matriz en cuatro partes iguales
A = T1[0:280,0:256]
B = T1[0:280,256:]
C = T1[280:,0:256]
D = T1[280:,256:]
print shape(A)
print shape(B)
print shape(C)
print shape(D)
```

```python
#Generamos una nueva matriz de zeros que recibe numeros complejos
T2 = np.zeros((560,512),dtype=complex)

#Llenamos la matriz con los valores de T1 pero en las nuevas posiciones
T2[0:280,0:256] = D
T2[0:280,256:] = C
T2[280:,0:256] = B
T2[280:,256:] = A
```

```python
#COOOOOOOOOOO
hist2d(T2[1].real, T2[0].real)
```

```python
#Convertimos en 0 todos los valores de la matriz T2 que se encuentran dentro de un circulo de 
#radio 100. centrado en el centro de la matriz, esto es un FILTRO PASA ALTAS
for j in range (560):
    for i in range (512):
        if (((i**2)-(256*i)+((-j)**2)+(280*j)) > -143836):
            T2[j,i] = 0.0 + 0.0j
```

```python
#Convertimos en 0 todos los valores de la matriz T2 que se encuentran en el centro, fuera de un circulo de radio 100
#centrado en el centro de la matriz, esto es un FILTRO PASA BAJAS
#for j in range (560):
 #   for i in range (512):
  #      if (((i**2)-(256*i)+((-j)**2)+(280*j)) > -133936):
   #         T5[j,i] = 0.0 + 0.0j
```

```python
hist2d(T2[1].real, T2[0].real)
colorbar()
```

```python
#Dividimos la matriz ya modificada nuevamente en cuatro partes iguales
E = T2[0:280,0:256]
F = T2[0:280,256:]
G = T2[280:,0:256]
H = T2[280:,256:]
```

```python
#Generamos una matriz de ceros y la llenamos con los valores de la matriz T2 pero ahora en la posiciones iniciales
T3 = np.zeros((560,512),dtype=complex)


T3[0:280,0:256] = H
T3[0:280,256:] = G
T3[280:,0:256] = F
T3[280:,256:] = E


print np.shape(T3)
```

```python
#Hacemos la transformada inversa de la matriz T3 O T2, para recuperar la nueva imagen
T4 = ifft2(T3,axes=(0,1)).real
print np.shape(T4)
```
