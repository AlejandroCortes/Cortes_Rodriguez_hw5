```python
#Generamos la transformada de Fourier de la señal y la normalizamos, esto es lo que se muestra en la siguiente grafica
fft_senal = fft(senal) / len(senal)
freq = fftfreq(len(senal), 1) 
plot(freq,np.abs(fft_senal), 'y')
plt.title('$TRANSFORMADA$'' ''$NORMALIZADA$', fontsize=15)
```

```python
#El espectro de potencias de la transformada de fourier, Tomado del Repositorio de Forero
half_n = np.ceil(len(senal))
fft_senal_half = (2.0) * fft_senal[:half_n]
freq_half = freq[:half_n]
#Graficas del Espectro de portencias
plt.figure(figsize=(16, 5))
plt.subplots_adjust(hspace=0.50, wspace=0.50)
plt.subplot(121)
scatter(freq_half, log10(np.absolute(fft_senal_half)**2), s=10) 
plt.title('$ESPECTRO$'' ''$DE$'' ''$POTENCIAS$', fontsize=15)
plt.subplot(122)
scatter(freq_half, log10(np.absolute(fft_senal_half)**2), s=10) 
plt.title('$ESPECTRO$'' ''$DE$'' ''$POTENCIAS$', fontsize=15)
plt.xlim([0,0.2])
plt.ylim([-20,-2])
```

### Ruido Blanco

El ruido blanco o sonido blanco es una señal aleatoria (proceso estocástico) que se caracteriza por el hecho de que sus valores de señal en dos tiempos diferentes no guardan correlación estadística. Como consecuencia de ello, su espectro de potencias(PSD) es una constante, es decir, su gráfica es plana. Esto significa que la señal contiene todas las frecuencias y todas ellas muestran la misma potencia.

Tomado de: http://es.wikipedia.org/wiki/Ruido_blanco

```python
#Espectro de potencias plano, para el ruido blanco
blanco =fft_senal_half/np.abs(fft_senal_half)
plot(np.absolute(blanco))
plt.title('$ESPECTRO$'' ''$PLANO$', fontsize=15)
```

```python
#Hacemos Transformafa inversa de Fourier, para recuperar el sonido del ruido blanco
blancoinver = ifft(blanco)
blancoreal = blancoinver.real
plot(blancoreal,'r')
plt.title('$RUIDO$'' ''$BLANCO$', fontsize=15)
```

```python
#Comparamos la señal inicial y el ruido blanco de la misma
plt.plot(senal, 'y')
plt.plot(blancoreal, 'r')
plt.xlabel('Eje x')
plt.ylabel('Eje Y')
plt.title('$SONIDO$'' ''$ORIGINAL$'' ''$Y$'' ''$RUIDO$'' ''$BLANCO$', fontsize=15)
```

```python
#Guardamos el rudio blanco en un archivo llamado "RuidoBlanco.wav"
ruidoblanco = audio.Sndfile("RuidoBlanco.wav", 'w', audio.Format('wav'),1, sampling)
ruidoblanco.write_frames(blancoreal)
ruidoblanco.close()
```

### Ruido Rosa

El Ruido rosa o el ruido de 1/frecuencia es una señal o proceso con un espectro de potencias tal que la densidad del espectro de potencia es inversamente proporcional a la frecuencia de la señal. En el ruido rosa, cada octava lleva una igual cantidad de ruido de potencia.

Tomado de: http://en.wikipedia.org/wiki/Pink_noise
