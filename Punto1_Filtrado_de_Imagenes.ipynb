## Filtrado de full_moon.jpg

```python
imagen1 = plt.imread('Imagenes/full_moon.jpg') #cargar imagen
print shape(imagen1) #Muestra las dimensiones de la imagen
plt.imshow(imagen1) #Muestra la imagen
```

Esta imagen tiene 296 pixeles de alto, 300 de largo y consta de 3 capas

```python
# usando la libreria de la Transformada de Fourier(TF) se halla la TF bidimensional para la imagen (señal)
T1 = fft2(imagen1) 
# imprime el tamaño de la matriz que es el resultado de la TF bidimensional
print np.shape(T1)
```

Con el fin de ver el comportamiento de la matríz que arroja la transformada se realiza lo siguiente:

1) Esta es la esquina superior izquierda del arreglo:

```python
a = np.zeros((5,5))
for i in range (5):
    for j in range (5):
        a[j,i]=np.abs(T1[j,i,0])

print a
        
```

2) Esta es la esquina inferior derecha del arreglo:

```python
a = np.zeros((5,5))
for i in range (295,300):
    for j in range (290,295):
        a[j-290,i-295]=np.abs(T1[j,i,0])

print a
```

3) Esta es la esquina inferior izquierda:

```python
a = np.zeros((5,5))
for i in range (5):
    for j in range (290,295):
        a[j-290,i]=np.abs(T1[j,i,0])

print a
```

4) Esta es la esquina superior derecha:

```python
a = np.zeros((5,5))
for i in range (295,300):
    for j in range (5):
        a[j,i-295]=np.abs(T1[j,i,0])

print a
```

ya que tenemos los valores máximos en las esquinas, vamos a modificar el arreglo de tal manera que ahora estos valores se encuentren en el centro del arreglo.

```python
A = T1[0:148,0:150]
B = T1[0:148,150:]
C = T1[148:,0:150]
D = T1[148:,150:]
print shape(A)
print shape(B)
print shape(C)
print shape(D)
```
