##### FILTROS PASA ALTAS Y PASA BAJAS EN IMAGENES

A continuación se realiza el filtrado de 4 señales bidimencionales (Imágenes) por medio de un filtro pasa bajas y otro pasa altas (frecuencias), usando la transformada de Fourier bidimensional en Python.

##### El procedimiento realizado se describe acontinuación:

1) Se carga la imagen incial y se mira el numero de capas que tiene cada imagen y de que tamaño son.

2) Se realiza la Transformada de fourier de la imagen.

3) Luego se organiza la matriz como indica la Figura 1, dejando el componente de frecuencia cero en el centro del espectro.

4) Para aplicar cada uno de los filtros seleccionamos un circulo con centro en la mitad de la matriz. Si el filtro es pasa altas los los valores que se encuentran ubicados dentro del circulo se convierten en 0. Pero, si el circulo es pasa bajas los valores que estan fuera del cirulo se convierten en 0.

5) Para regresar a la nueva imagen tenemos que volver a organizar la matriz a como estaba antes del punto 3.

6) Aplicamos Transformada Inversa de Fourier a la matriz del punto 5.

7) Graficamos la imagen final.

## FILTROS DE LA IMAGEN:  20_popc_cho009-1.tif

#### 1)

```python
#Carga y muestra la imagen 20_popc_cho009-1.tif
imagen1 = pylab.imread("Imagenes/20_popc_cho009-1.tif")
fig = figure(1, figsize=(8,8))
imshow(imagen1[:,:], cmap = cm.Greys_r)
plt.title('$IMAGEN$'' ''$ORIGINAL$', fontsize=20)
```

### FILTRO PASA ALTAS

#### 2)

```python
#Usando la libreria de la Transformada de Fourier(TF) se halla la TF bidimensional 
#para la imagen (señal)
T1 = fft2(imagen1) 
#Imprime el tamaño de la matriz que es el resultado de la TF bidimensional
print np.shape(T1)
```

Esta imagen tiene 560 pixeles de alto, 512 de largo y consta de 1 capa

#### 3)

```python
#Organiza la matriz dejando el componente de cero frecuencia en el centro.
T2 = fftshift(T1)
```

#### 4)

```python
#Convertimos en 0 todos los valores de la matriz T2 que se encuentran dentro de un circulo de 
#radio 20. centrado en el centro de la matriz (256,560), esto es un FILTRO PASA ALTAS
for j in range (560):
    for i in range (512):
        if ((i**2)-(512*i)+(j**2)-(560*j)) < -143511:
            T2[j,i] = 0
```

#### 5)

```python
#Se reorganiza la matriz a sus posiciones iniciales
T3 = ifftshift(T2)
```

#### 6)

```python
#Hacemos la transformada inversa de la matriz T3, para recuperar la nueva imagen
T4 = ifft2(T3).real
```

#### 7)

```python
#Ploteamos la nuevas imagenes despues de pasar por los filtros
fig = plt.figure(1,figsize=(8,8))
plt.imshow(T4, cmap = cm.Greys_r)
plt.title('$FILTRO$'' ''$PASA$'' ''$ALTAS$',fontsize=20)
```

### FILTRO PASA BAJAS

#### 2)

```python
#Usando la libreria de la Transformada de Fourier(TF) se halla la TF bidimensional 
#para la imagen (señal)
Q1 = fft2(imagen1) 
#Imprime el tamaño de la matriz que es el resultado de la TF bidimensional
print np.shape(Q1)
```

#### 3)

```python
#Organiza la matriz dejando el componente de cero frecuencia en el centro.
Q2 = fftshift(Q1)
```

#### 4)

```python
#Convertimos en 0 todos los valores de la matriz Q2 que se encuentran fuera de un circulo de 
#radio 20. centrado en el centro de la matriz, esto es un FILTRO PASA BAJAS
for j in range (560):
    for i in range (512):
        if ((i**2)-(512*i)+(j**2)-(560*j)) > -143511:
            Q2[j,i] = 0
```

#### 5)

```python
#Se reorganiza la matriz a sus posiciones iniciales
Q3 = ifftshift(Q2)
```

#### 6)

```python
#Hacemos la transformada inversa de la matriz Q3, para recuperar la nueva imagen
Q4 = ifft2(Q3).real
```

#### 7)

```python
#Ploteamos las nuevas imagenes despues de pasar por los filtros
fig = plt.figure(1,figsize=(8,8))
plt.imshow(Q4, cmap = cm.Greys_r)
plt.title('$FILTRO$'' ''$PASA$'' ''$BAJAS$',fontsize=20)
```
